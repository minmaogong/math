```python
import numpy as np
```

## 矩阵及其运算

```python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,3,7],[5,0,2]])
```

```python
print(A)
print(B)
print(A.shape)
```

```python
# 1. 转置
print(A.T)
print(A.T.shape)
```

```python
# 2. 矩阵相加
print(A + B)
```

```python
print((A + B).T)
print(A.T + B.T)
```

```python
# 3. 矩阵与标量相加
print(A + 5)
```

```python
# 4. 矩阵与向量相加
x = np.array([2,4,6])
print(A + x)
```

```python
# 5. 矩阵与标量相乘
print(A * 3)
```

```python
# 6. 矩阵与向量相乘
print(A * x)
```

```python
# 7. 矩阵与矩阵相乘(阿达玛积，两个矩阵对应元素乘积，计算出的矩阵形状不变)
print(A * B)
```

```python
# 8. 矩阵乘法（点积）
print(A.dot(B.T))
```

```python
print(A @ B.T) # 与 A.dot(B.T) 一样，也是矩阵乘法
```

```python
print(B.T @ A)
```

```python
print((A @ B.T).T)
```

```python
print(B @ A.T)
```

```python
# 9. 矩阵求逆
C = np.array([[1,2],[3,5]])
print(C)
```

```python
C_inv = np.linalg.inv(C) # 方阵求逆函数
print(C_inv)
```

```python
print(C @ C_inv)
print(C_inv @ C)
```

```python
C_inv = np.around(C_inv).astype(int)
print(C_inv)
```

```python
print(C @ C_inv)
print(C_inv @ C)
```

```python
A_inv = np.linalg.pinv(A) # 伪逆
print(A_inv)
```

```python
print(A @ A_inv) # 矩阵与伪逆点积计算的结果是单位矩阵
```

```python
print(A_inv @ A) # 伪逆与矩阵点积计算的结果不是单位矩阵
```

```python
# 10. 获取矩阵元素
print(A[0][1])
```

```python
print(A[0,1])
```

```python
# 11. 矩阵向量化
print(A.flatten()) # 行向量化
```

```python
# 12. 矩阵内积 （两个矩阵对应元素乘积之和。是一个标量）
# 利用阿达玛积 （两个矩阵各位置元素相乘，形状不变）
print(A * B)
print(np.sum(A * B))
```

```python
# 利用向量化
print(np.dot(A.flatten(), B.flatten())) # np.dot(A.flatten(), B.flatten()) 两个向量作点积运算，即各位置元素对应相乘，再求和
```

```python
# 13. 矩阵克罗内克积（张量积）
print(np.kron(A, B))
```

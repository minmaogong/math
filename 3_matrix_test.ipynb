```python
import numpy as np
```

```python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,3,7],[5,0,2]])
```

```python
print(A)
print(A.shape)
```

```python
# 1. 转置
print(A.T)
print(A.T.shape)
```

```python
# 2. 矩阵相加
print(A + B)
```

```python
print((A + B).T)
print(A.T + B.T)
```

```python
# 3. 矩阵与标量相加
print(A + 5)
```

```python
# 4. 矩阵与向量相加
x = np.array([2,4,6])
print(A + x)
```

```python
# 5. 矩阵与标量相乘
print(A * 3)
```

```python
# 6. 矩阵与向量相乘
print(A * x)
```

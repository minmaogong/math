```python
import numpy as np
```

## 均匀分布

```python
# 1. [0, 1) 分布的小数
a = np.random.rand(5) # 向量
print(a)
```

```python
a = np.random.rand(2, 3) * 10 + 10 # 矩阵，每个元素在 [10, 20) 之间
print(a)
```

```python
# 2. 整数
b = np.random.randint(0, 100, size=[2, 3])
print(b)
```

## 正态分布

```python
# 1. 标准正态分布
c = np.random.randn(10) # 10个数构成的向量，每一个元素服从标准正态分布。均值为0，方差为1，都集中在0附近，离0越近，可能性越大
print(c)
```

```python
c = np.random.randn(10) * 10 + 5 # 以 5 为均值，方差是10，一般正太分布，分布较广
print(c)
```

```python
# 2. 一般正态分布
d = np.random.normal(5, 10, size=(2, 3)) # 均值=5，方差=10
print(d)
```

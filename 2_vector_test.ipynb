```python
import numpy as np
```

## 向量的基本运算

```python
# 定义一个向量
x = np.array([2, 5, 8])
print(x)
```

```python
print(x.shape)
```

```python
# 1. 向量转置
print(x.T)
print(x.T.shape)
```

```python
# 2. 向量相加
y = np.array([1, 3, 7])
print(x + y)
```

```python
print(x + 5)
```

```python
# 3. 向量相乘
# 与标量相乘
print(x * 3)
```

```python
# 与向量相乘
print(x * y)
```

```python
# 4. 向量点积（内积）
print(x.dot(y))
```

## 向量范数

```python
# L0范数
norm_l0 = np.linalg.norm(x, ord=0)
print(norm_l0)
```

```python
# L1范数
norm_l1 = np.linalg.norm(x, ord=1)
print(norm_l1)
```

```python
# L2范数
norm_l2 = np.linalg.norm(x, ord=2)
print(norm_l2)
```

```python
# 无穷范数
norm_inf = np.linalg.norm(x, ord=np.inf)
print(norm_inf)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 绘制y=sinx的图像

```python
# 定义自变量x
x = np.arange(0, 6, 0.1)
```

```python
# 计算因变量y
y = np.sin(x)
```

```python
print(x)
print(y)
```

```python
# 绘制图形
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=sin(x)")
plt.show()
```

## 绘制导函数图像

```python
# 计算导函数
y1 = np.cos(x)
```

```python
plt.plot(x, y1)
plt.xlabel("x")
plt.ylabel("y1")
plt.title("y1=cos(x)")
plt.show()
```

## 放在一个图中显示

```python
# 绘制图像
plt.plot(x, y, label="y=sin(x)")
plt.plot(x, y1, label="y=sin'(x)", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=sin(x)")
plt.legend() # 显示图例
plt.show()
```
